```python
import pandas as pd

# 假设 df2 已经定义并包含 '教育类型'、'type_work' 和 '最高学历' 列

# 创建一个空的 DataFrame，用于存储所有工作类别的分析结果
all_degree_type_df = pd.DataFrame()
all_type_degree_df = pd.DataFrame()

# 按照 "11类_work" 这一列将 df2 分成若干个子数据框
for work_category, group in df2.groupby('11类_work'):
    print(f"Processing work category: {work_category}")
    
    # 1. 统计在相同“最高学历”人群里，不同“type_work”的频率和占比
    degree_type_results = []
    for degree, sub_group in group.groupby('最高学历'):
        # 统计 type_work 的频率
        work_counts = sub_group['type_work'].value_counts().reset_index()
        work_counts.columns = ['type_work', '计数']
        
        # 计算总计数
        total_count = work_counts['计数'].sum()
        
        # 计算占比
        work_counts['占比'] = work_counts['计数'] / total_count
        
        # 添加“最高学历”列
        work_counts['最高学历'] = degree
        
        # 添加“11类_work”列
        work_counts['11类_work'] = work_category
        
        # 将结果添加到列表
        degree_type_results.append(work_counts)

    # 合并所有结果
    degree_type_df = pd.concat(degree_type_results, ignore_index=True)
    all_degree_type_df = pd.concat([all_degree_type_df, degree_type_df], ignore_index=True)

    # 2. 统计在相同“type_work”人群里，不同“最高学历”的频率和占比
    type_degree_results = []
    for degree, sub_group in group.groupby('最高学历'):
        # 统计 type_work 的频率
        degree_counts = sub_group['最高学历'].value_counts().reset_index()
        degree_counts.columns = ['最高学历', '计数']
        
        # 计算总计数
        total_count = degree_counts['计数'].sum()
        
        # 计算占比
        degree_counts['占比'] = degree_counts['计数'] / total_count
        
        # 添加“type_work”列
        degree_counts['type_work'] = work_category
        
        # 添加“11类_work”列
        degree_counts['11类_work'] = work_category
        
        # 将结果添加到列表
        type_degree_results.append(degree_counts)

    # 合并所有结果
    type_degree_df = pd.concat(type_degree_results, ignore_index=True)
    all_type_degree_df = pd.concat([all_type_degree_df, type_degree_df], ignore_index=True)

# 设置文件路径和文件名
output_folder = r"D:\songyue\精英大学分析\分学科"
output_filename = "all_degree_type_analysis.xlsx"
output_path = f"{output_folder}\\{output_filename}"

# 将 all_degree_type_df 和 all_type_degree_df 输出到同一个 Excel 表格的不同工作表中
with pd.ExcelWriter(output_path) as writer:
    # 输出第一个 DataFrame
    all_degree_type_df.to_excel(writer, sheet_name="Degree_Type_Frequency", index=False)
    # 输出第二个 DataFrame
    all_type_degree_df.to_excel(writer, sheet_name="Type_Degree_Frequency", index=False)

print(f"All results saved to {output_path}")
```

```python
import pandas as pd

# 假设 final_merged_df 已经定义

# 定义要应用的筛选条件列表
conditions = [
    ("university", df2['type_work'] == 'university'),
    ("industry", df2['type_work'] == 'industry'),
    ("government", df2['type_work'] == 'government'),
    ("Labs", df2['type_work'] == 'Labs'),
    # ("government_nonprofit", (df2['type_work'] == 'government') & (df2['type_work'] == 'nonprofit'))
    ("government_or_nonprofit", (df2['type_work'] == 'government') | (df2['type_work'] == 'nonprofit'))
]

# 定义用于存储所有条件下结果的列表
all_results = []

# 遍历条件并应用过滤
for condition_name, condition_filter in conditions:
    # 按条件筛选数据
    filtered_df = df2[condition_filter & (df2['教育当前工作'] == 1)]
    
    # 定义用于存储每个 condition_name 下结果的列表
    results = []
    
    # 根据“教育类型”分组
    for edu_type, group in filtered_df.groupby('教育类型'):
        # 统计国家唯一ID_work频率，并取前100个国家
        top_countries = group['国家唯一ID_work'].value_counts().head(100).index.tolist()
        
        # 遍历前100个国家，获取每个国家的数据
        for country in top_countries:
            # 筛选出属于该国家的样本
            country_samples = group[group['国家唯一ID_work'] == country]
            
            # 统计唯一ID_edu的频率
            id_counts = country_samples['唯一ID_edu'].value_counts().reset_index()
            id_counts.columns = ['唯一ID_edu', '计数']
            
            # 去掉计数小于（含）3的记录
            id_counts = id_counts[id_counts['计数'] >= 4]
            
            # 计算剩余记录的总计数
            valid_total_count = id_counts['计数'].sum()
            
            # 如果没有满足条件的记录，则跳过
            if valid_total_count == 0:
                continue
            
            # 重新计算占比
            id_counts['占比'] = id_counts['计数'] / valid_total_count
            
            # 添加 o_id 列、教育类型和国家字段
            id_counts['o_id'] = id_counts['唯一ID_edu']
            id_counts['教育类型'] = edu_type
            id_counts['国家'] = country
            id_counts['condition'] = condition_name  # 添加当前条件名称
            
            # 将结果添加到列表
            results.append(id_counts)
    
    # 将每个条件下的结果合并
    if results:
        condition_result_df = pd.concat(results, ignore_index=True)
        all_results.append(condition_result_df)

# 合并所有条件的结果
final_result_df = pd.concat(all_results, ignore_index=True)

# 导出到 CSV 文件
output_path = r"D:\songyue\精英大学分析\基尼系数计算准备df——国内多条件.csv"
final_result_df.to_csv(output_path, index=False)

print("结果已保存到:", output_path)
```

```python
import pandas as pd

# 假设df是包含数据的DataFrame
# 按教育类型、国家和条件分组
grouped = final_result_df.groupby(['教育类型', '国家', 'condition'])

# 定义用于存储结果的列表
results = []

# 遍历每个组
for (edu_type, country, condition), group in grouped:
    # 按占比降序排列
    group = group.sort_values('占比', ascending=False)
    # 计算累计占比
    group['累计占比'] = group['占比'].cumsum()
    # 找到累计占比刚好超过0.8的行
    threshold_index = group[group['累计占比'] > 0.2].index[0]
    # 统计唯一ID_edu数量
    unique_id_count = group.loc[:threshold_index, '唯一ID_edu'].nunique()
    # 计算该组总的唯一ID_edu个数
    total_unique_id_count = group['唯一ID_edu'].nunique()
    # 计算百分比
    percentage = unique_id_count / total_unique_id_count * 100
    # 将结果添加到列表
    results.append({
        '教育类型': edu_type,
        '国家': country,
        'condition': condition,
        '累计占比超过0.8的唯一ID_edu个数': unique_id_count,
        '占总唯一ID_edu个数的百分比': percentage,
        '总的唯一ID_edu个数': total_unique_id_count  # 新增列
    })
# 转换为DataFrame
result_df2 = pd.DataFrame(results)
# 将 result_df 保存为 CSV 文件
output_path = r"D:\songyue\精英大学分析\基尼系数计算准备df——国内多条件-20%二八法则.csv"
result_df2.to_csv(output_path, index=False)
```

```python
# 函数：使用 IQR 剔除异常值再计算分组相关系数的示例：
import pandas as pd
# 假设 gini_result_df 已经定义，并包含 '基尼系数'、'唯一ID_edu个数' 和 '教育类型' 列

# 函数：使用 IQR 剔除异常值
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 * IQR)]

# 存储每个教育类型分组的相关系数结果
robust_correlation_results = []

# 按 "教育类型" 分组，并计算去除异常值后的相关系数
for edu_type, group in gini_result_df.groupby('教育类型'):
    # 去除异常值
    group_no_outliers = remove_outliers_iqr(group, '基尼系数')
    group_no_outliers = remove_outliers_iqr(group_no_outliers, '唯一ID_edu个数')
    
    # 计算相关系数
    correlation = group_no_outliers[['基尼系数', '唯一ID_edu个数']].corr(method='spearman').iloc[0, 1]
    robust_correlation_results.append({
        '教育类型': edu_type,
        'Gini-唯一ID_edu相关系数（去除异常值后）': correlation
    })

# 转换为 DataFrame 并显示结果
robust_correlation_df = pd.DataFrame(robust_correlation_results)
print(robust_correlation_df)
# 保存为 CSV 文件（可选）
output_path = r"D:\songyue\精英大学分析\分组基尼系数与唯一ID_edu相关系数（去除异常值）.csv"
robust_correlation_df.to_csv(output_path, index=False)
print(f"结果已保存到: {output_path}")
```

```python
import pandas as pd

# 假设 data 是包含数据的原始 DataFrame
# 按“3类_work”、“教育类型”和“国家”组合进行分组
grouped = final_result_df.groupby(['排名2007', '教育类型', '国家'])

# 定义用于存储每个子df基尼系数的列表
gini_results = []

# 遍历每个子 DataFrame
for (work_category, edu_type, country), df in grouped:
    # 检查子数据框是否为空
    if df.empty:
        continue

    # 计算唯一ID_edu的行数
    unique_id_count = len(df['唯一ID_edu'])
    
    # 计算 x 占比和 x 累计占比
    df['x占比'] = 1 / unique_id_count
    df['x累计占比'] = df['x占比'].cumsum()
    
    # 计算 y 占比和 y 累计占比
    total_count = df['计数'].sum()
    if total_count == 0:
        # 如果总计数为0，则跳过该分组
        continue
    df['y占比'] = df['计数'] / total_count
    df['y累计占比'] = df['y占比'].cumsum()
    
    # 计算梯形面积
    df['梯形面积'] = (df['x累计占比'].diff() * (df['y累计占比'] + df['y累计占比'].shift(fill_value=0))) / 2
    df.at[0, '梯形面积'] = (df['x累计占比'].iloc[0] * df['y累计占比'].iloc[0]) / 2
    
    # 重置索引，确保索引从0开始
    df = df.reset_index(drop=True)

    # 计算梯形面积之和并赋值到第一行
    df.at[0, '梯形面积之和'] = df['梯形面积'].sum()

    # 计算基尼系数并赋值到第一行
    df.at[0, '基尼系数'] = 2 * df.at[0, '梯形面积之和'] - 1
    
    # 提取当前子 DataFrame 的基尼系数的第一个值
    gini_value = df['基尼系数'].iloc[0]
    # 检查基尼系数是否为 NaN
    if pd.notna(gini_value):
        gini_results.append({
            '排名2007': work_category,
            '教育类型': edu_type,
            '国家': country,
            '基尼系数': gini_value,
            '唯一ID_edu个数': unique_id_count  # 添加唯一ID_edu的行数
        })

# 将结果转换为 DataFrame
gini_result_df = pd.DataFrame(gini_results)

# 将结果保存为 CSV 文件
output_path = r"D:\songyue\精英大学分析\基尼系数计算准备df——国内大学2-分组基尼系数汇总-排名2007.csv"
gini_result_df.to_csv(output_path, index=False)

print("结果已保存到:", output_path)
```
